```python
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
import tkinter as tk
import matplotlib.colors as mcolors

```

```python
shp_file = '/Users/rebeccawagner/Documents/GitHub/GerryGainMCMC/Data/Alabama shp files/Alabama_VTD_District_Intersection.shp'
gdf = gpd.read_file(shp_file)
```

```python
gdf["geometry"][468]
```

```python
gdf["incumbent"] = gdf["incumbent"].fillna(0)
gdf
```

```python
gdf['centroid'] = gdf['geometry'].centroid
gdf["centroid_x"] = gdf['geometry'].centroid.x
gdf["centroid_y"] = gdf['geometry'].centroid.y
```

```python
gdf.head()
```

```python
fig, ax = plt.subplots(1,1,figsize=(15,15))
ax.axis('off')

for i in range(len(gdf)):
    polygon = gdf['geometry'][i]
    x, y = polygon.exterior.xy
    ax.fill(x,y,picker=True)
```

```python
fig, ax = plt.subplots(1,1,figsize=(30,30))
gdf.plot(ax=ax)
ax.scatter(gdf['centroid_x'], gdf['centroid_y'], s=2, color='k', alpha=.5, picker=True)

ax.axis('off')

plt.show()

def on_pick(event):
   print(event)

cid = fig.canvas.mpl_connect('pick_event', on_pick)
```

```python
collection = ax.collections[0]
```

```python
n = len(collection.get_paths())
facecolors = collection.get_facecolors()
```

```python
if len(facecolors) == 1 and n != 1:
    facecolors = np.array([facecolors[0]] * n)

facecolors[3] = mcolors.to_rgba("crimson")
collection.set_facecolor(facecolors)

plt.show()
```
