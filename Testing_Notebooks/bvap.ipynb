```python
file = '/Users/rebeccawagner/Documents/GitHub/GerryGainMCMC/Output/Chain_05.02/used/RunTwo_assignments_1_2023-05-04-01-09-18-132730'
vtd_data = '/Users/rebeccawagner/Documents/GitHub/GerryGainMCMC/Data/bvap_vtds.csv'
seed_hist = '1'

def bvap_output(file, vtd_data, seed_hist):

    bvap_scores = list()

    data = pd.read_csv(file)
    vtd_data = pd.read_csv(vtd_data)
    districts = data['1'].unique()
    data = data.set_index('Iteration')
    data.index = data.index.map(lambda x: x[:-6])

    for column in data.columns:
        score_list = []
        for district in districts:
            vtds = list(data[data[column] == district].index)
            district_data = vtd_data[vtd_data['block_vtda'].isin(vtds)]
            bvap = district_data.sum()['black_combo_18']/district_data.sum()['total_18'] 
            score_list.append(bvap)
        bvap_scores.append(score_list)

    frame = pd.DataFrame(bvap_scores, columns=districts)
    return frame
```

```python
test = bvap_output(file, vtd_data, seed_hist)
path = '/Users/rebeccawagner/Documents/GitHub/GerryGainMCMC/Output/Chain_05.02/bvap' + '/' + 'RunTwo_bvap_' + '1'

with open(path, 'x', encoding='utf8', newline='') as output_file:
            test.to_csv(output_file, sep=',', encoding='utf-8', header=True, index=False)

```

```python
demo_data = pd.read_csv('/Users/rebeccawagner/Documents/GitHub/GerryGainMCMC/Data/bvap_data')
directory = glob('/Users/rebeccawagner/Documents/GitHub/GerryGainMCMC/Output/Chain_02.23/used' + '/' +'*') + glob('/Users/rebeccawagner/Documents/GitHub/GerryGainMCMC/Output/Chain_02.23/next' + '/' +'*')

output_path = '/Users/rebeccawagner/Documents/GitHub/GerryGainMCMC/Output/Chain_02.23/bvap'

districts = [1001,1002,1003,1004,1005,1006,1007]

for file in directory:

    bvap_output = list()

    seed = file.split('_')[3]

    assignments = pd.read_csv(file)
    assignments = assignments.set_index('Iteration')
    assignments.index = assignments.index.map(lambda x: x[:-6])

    for i,column in enumerate(assignments.columns):
        score_list = []
        for district in districts:
            vtds = list(assignments[assignments[column] == district].index)
            district_data = demo_data[demo_data['block_vtda'].isin(vtds)]
            bvap = district_data.sum()['black_combo_18']/district_data.sum()['total_18'] 
            score_list.append(bvap)
        bvap_output.append(score_list)

    frame = pd.DataFrame(bvap_output, columns=districts)
    path = output_path + '/' + 'RunOne_bvap_' + seed

    try:
        with open(path, 'x', encoding='utf8', newline='') as output_file:
            frame.to_csv(output_file, sep=',', encoding='utf-8', header=True, index=False)
    except FileExistsError:

        print('dup')
        path = path + '_dup'
        
        with open(path, 'x', encoding='utf8', newline='') as output_file:
            frame.to_csv(output_file, sep=',', encoding='utf-8', header=True, index=False)

```
