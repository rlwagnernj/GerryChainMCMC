```python
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime as dt
from glob import glob
import os

EXT = '.rebecca'

def TimeStamp(seed):
    return str(dt.now()).replace(':','-').replace(' ', '_') + '-' + str(seed)

def Move(pos):
    pos+=np.random.normal(scale=0.1, size=pos.shape)
    return

def Save(pos, seed):
    file_name = TimeStamp(seed) + EXT
    np.savetxt(file_name, pos)
    #with open(file_name, 'w') as f:
    #    f.write(str(pos))
        
def RunBranch(seed, pos):
    np.random.seed(seed)
    steps = 1000
    for _ in range(steps):
        Move(pos)
    Save(pos, seed)

def GetFiles():
    files = glob('*' + EXT)
    return files
```

```python
# Initialize first run
N = 100
pos = np.random.uniform(low=-1, high=1, size=(N,2))

seed = 101
RunBranch(seed, pos)
```
